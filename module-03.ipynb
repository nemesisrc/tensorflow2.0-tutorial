```python
# data points
x = [1, 2, 2.5, 3, 4]
y = [1, 4, 7, 9, 15]
```

```python
# creating the plot
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression with TensorFlow')
plt.plot(x, y, 'bs', label='Data Points')
plt.legend()
```

```python
# creating the regression/best-fit line (y = mx + c)

# get the line coefficients(slope and intercept)
coefficients = np.polyfit(x, y, 1)   
polynomial = np.poly1d(coefficients)

# create smooth X points for the line
x_smooth = np.unique(x)
y_smooth = polynomial(x_smooth)

# plot the best-fit/regression line
plt.plot(x_smooth, y_smooth, 'r--', label='Best-Fit/Regression Line')
plt.legend()
```
